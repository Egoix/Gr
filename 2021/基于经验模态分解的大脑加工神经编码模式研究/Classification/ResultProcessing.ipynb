对结果进行处理,计算average,并写入csv文件最后一行

```python
import pandas as pd
import numpy as np
import csv
import os
```

```python
def csv_delimiter_fixer(filename):
    with open(filename, 'a+b') as fileobj:
        fileobj.seek(-1, 2)
        if fileobj.read(1) != b"\n":
            fileobj.write(b"\r\n")
```

```python
data_path = r'D:\project\BrainEEG\202105\qian_Result_other_c_2021-05-17'     # file fold path
for data_file_name in os.listdir(data_path):
    print(data_path,data_file_name)
    # read file
    csv_data = pd.read_csv(f'{data_path}\\{data_file_name}', encoding = "ISO-8859-1", index_col=0) # header = 1
    # get columns  name
    data_columns = csv_data.columns
    avg = ['avg']
#     for column in data_columns:
# 由于最后一列combination没有最后一列，因此使用data_columns[:-1]
    for column in data_columns[:-1]:
        # split and get data
        csv_data[column+'_split_1'], csv_data[column+'_split_2'] = csv_data[column].str.split('%¡À').str
        remove0 = lambda m:m.group(0).rstrip("0")
        csv_data[column+'_split_2'].replace(regex=True,inplace=True,to_replace=r'%',value=r'')
        average = csv_data[column+'_split_1'].astype(float).mean()
        std = csv_data[column+'_split_2'].astype(float).mean()
        average = '%.2f' % average
        std = '%.2f' % std
        pieces=[average,'%±',std,'%']
        # print(pieces)
        pieces = ''.join(pieces)
        # print(''.join(pieces))
        avg.append(pieces)
    # write average into file
    filename = f'{data_path}\\{data_file_name}'
    csv_delimiter_fixer(filename)
    with open(f'{data_path}\\{data_file_name}', 'a', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(avg)
```
