```python

import mne
import numpy as np
import scipy.io
import matplotlib.pyplot as plt
import pandas as pd 
import matplotlib.patches as mpathes

def cal_ctm(signal: np.ndarray, r: float, tao=1) -> float:
    point_num = 0

    center_x = np.mean(signal)
    n = len(signal)

    for i in range(n):
        tmp_point = (signal[i], signal[(i + tao) % n])
        point_num += 1 if (tmp_point[0] - center_x) ** 2 + (tmp_point[1] - center_x) ** 2 <= r * r else 0

    return point_num / n

def calculate_ctm(r):
    for xd in range(1,16):
        for yd in range(1,4):
            epochs_data = mne.read_epochs(f'D:\数据集\epochs_SEED//{xd}_{yd}-epo.fif')

            events_id_dict = epochs_data.event_id
            count_events = {key: 0 for key in events_id_dict.keys()}

            ans = np.zeros((len(epochs_data), 62))

            df = pd.DataFrame()

            for i in range(len(epochs_data)):
                trail_epoch = epochs_data[i]
                trail_channel_name = trail_epoch.ch_names
                trail_event = trail_epoch.event_id
                # df[trail_event] = []

                data = trail_epoch.get_data()
                for j in range(62):
                    # 1e6 是 微伏 单位
                    arr=data[0,j,:]
                    for t in range(len(arr)):
                        arr1=[arr[t] for t in range(37001)]
                        break
                    arr1=np.array(arr1)
                    #print(type(arr1))
                    ctm = cal_ctm(arr1*0.1, r)
                    ans[i][j] = ctm

                np.savetxt(f'{xd}_{yd}_r={r}_.csv', ans, delimiter=',')

def draw_psr():
    #画psr图
    #读取epochs文件
     for xd in range(1,16):
        for yd in range(1,4):
            epochs_data = mne.read_epochs(f'D:\数据集\epochs_SEED//{xd}_{yd}-epo.fif')
            events_id_dict = epochs_data.event_id

            print(epochs_data)
            #print(events_id_dict)
           
            for eid in events_id_dict.keys():
                #对某一通道进行操作
                channels_list = ['']
                trails_data = epochs_data.pick_channels(channels_list)
                #print(trails_data)
                #取平均值
                trails_data = trails_data[eid].average()
                #画出平均值图
                #trails_data.plot(time_unit='s')
                #单位问题
                data = trails_data.data*0.1    
                      
                for channel_id in range(len(channels_list)):
                    signal_data = data[channel_id]
                    #print(events_id_dict[eid] )
                    #print(channel_id )
                    
                    
                    #去0
                    #初始你想要操作的元组
                    arr = signal_data
                    #遍历列表
                    #37001是最小的长度
                    for i in range(len(arr)):
                        arr1=[arr[i] for i in range(37001)]
                        break
                    #print(len(arr1)) 
                 
                plt.subplot(2, 2, events_id_dict[eid] + channel_id * 4)
               # plt.contour(q1, q2, q1**2 + q2**2, [1])
                plt.plot(arr1, np.append(arr1[1:], arr1[0]), linewidth=0.25)
                #plt.plot(circle, linewidth=0.25)
                if events_id_dict[eid] + channel_id * 4 <= 4:
                    plt.title(f"{eid}")

                if eid == 'negative' and channel_id == 0:
                    plt.xlabel('μV')
                    plt.ylabel('μV')

                if eid == 'negative':
                    plt.ylabel(f"{channels_list[channel_id]}\n μV")
                plt.xlim(-20, 20)
                plt.ylim(-20, 20)

            #plt.suptitle(f"person:{xd}_{yd}_Fpz")
            filename = f"{xd}_{yd}Fpz_psr.png"
            #plt.savefig(filename, dpi=1000)
            plt.show()
            plt.close()



if __name__ == '__main__':
    #计算ctm 特征
    #calculate_ctm(1)
    #calculate_ctm(2)
    #calculate_ctm(10)


    # 画psr图
    draw_psr()
```
